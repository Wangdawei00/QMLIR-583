```python
import pyzx as zx
import sympy
```

```python
# Uses the following identity:
#         ───W(a)───W(b)───
#         ≡ ───Z^-a───X───Z^a───Z^-b───X───Z^b───
#         ≡ ───Z^-a───Z^-a───Z^b───X───X───Z^b───
#         ≡ ───Z^-a───Z^-a───Z^b───Z^b───
#         ≡ ───Z^2(b-a)───

# ──W(a)─ ≡ ───Z^-a───X───Z^a─
def W(a):
    c = zx.Circuit(qubit_amount=1)
    c.add_gate("ZPhase", 0, "-" + a)
    c.add_gate("NOT", 0)
    c.add_gate("ZPhase", 0, a)
    return c
```

```python
# zx.draw(W("1/7"))

c = W("1/7") + W("1/3")
zx.draw(c)
g = c.to_graph()
zx.full_reduce(g, quiet=False)
zx.draw(g)
```
